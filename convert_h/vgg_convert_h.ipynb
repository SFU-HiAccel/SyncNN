```python
from keras.models import load_model
import scipy.io as sio
import hdf5storage as io
import numpy
model = load_model('/content/cifar10_vgg13_da_9217.h5')
i=0
for layer in model.layers:
     i= i+1
     if len(layer.weights) > 0:
        print(layer.name,i, layer.weights[0].shape)


weights={}
weights['wc1'] = (model.layers[0].weights[0]).numpy()
weights['wc2'] = (model.layers[2].weights[0]).numpy()
weights['wc3'] = (model.layers[4].weights[0]).numpy()
weights['wc4'] = (model.layers[6].weights[0]).numpy()
weights['wc5'] = (model.layers[8].weights[0]).numpy()
weights['wc6'] = (model.layers[10].weights[0]).numpy()
weights['wc7'] = (model.layers[13].weights[0]).numpy()
weights['wc8'] = (model.layers[15].weights[0]).numpy()
weights['wc9'] = (model.layers[18].weights[0]).numpy()
weights['wc10'] = (model.layers[20].weights[0]).numpy()
weights['wd1'] = (model.layers[24].weights[0]).numpy()
weights['wd2'] = (model.layers[25].weights[0]).numpy()
weights['wd3'] = (model.layers[26].weights[0]).numpy()

io.savemat('weightsc1', {'wc1': weights['wc1']})
io.savemat('weightsc2', {'wc2': weights['wc2']})
io.savemat('weightsc3', {'wc3': weights['wc3']})
io.savemat('weightsc4', {'wc4': weights['wc4']})
io.savemat('weightsc5', {'wc5': weights['wc5']})
io.savemat('weightsc6', {'wc6': weights['wc6']})
io.savemat('weightsc7', {'wc7': weights['wc7']})
io.savemat('weightsc8', {'wc8': weights['wc8']})
io.savemat('weightsc9', {'wc9': weights['wc9']})
io.savemat('weightsc10', {'wc10': weights['wc10']})
io.savemat('weightsd1', {'wd1': weights['wd1']})
io.savemat('weightsd2', {'wd2': weights['wd2']})
io.savemat('weightsd3', {'wd3': weights['wd3']})
```
